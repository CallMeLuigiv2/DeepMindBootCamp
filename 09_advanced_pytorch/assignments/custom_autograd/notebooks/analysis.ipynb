## Part 1: Parameterized Swish Visualization

```python
# Visualize the Parameterized Swish for different beta values
x = torch.linspace(-5, 5, 200)
betas = [0.5, 1.0, 1.5, 2.0, 5.0]

plt.figure(figsize=(10, 5))
for beta_val in betas:
    beta = torch.tensor(beta_val)
    y = ParameterizedSwish.apply(x, beta)
    plt.plot(x.numpy(), y.detach().numpy(), label=f'beta={beta_val}')

plt.xlabel('x')
plt.ylabel('f(x) = x * sigmoid(beta * x)')
plt.title('Parameterized Swish for Different Beta Values')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
