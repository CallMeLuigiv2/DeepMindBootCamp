## 1. Load TensorBoard Logs

You can load TensorBoard event files directly, or just reference
your TensorBoard screenshots here.

```python
# Option 1: Load from TensorBoard event files
try:
    from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

    ea = EventAccumulator('../runs/')
    ea.Reload()

    # Extract scalars
    train_loss = [(s.step, s.value) for s in ea.Scalars('Loss/train')]
    val_loss = [(s.step, s.value) for s in ea.Scalars('Loss/val')]
    train_acc = [(s.step, s.value) for s in ea.Scalars('Accuracy/train')]
    val_acc = [(s.step, s.value) for s in ea.Scalars('Accuracy/val')]
    lr_history = [(s.step, s.value) for s in ea.Scalars('LR')]

    print(f'Loaded {len(train_loss)} epochs of training data')
except Exception as e:
    print(f'Could not load TensorBoard logs: {e}')
    print('Train your model first, or manually enter results below.')
```
