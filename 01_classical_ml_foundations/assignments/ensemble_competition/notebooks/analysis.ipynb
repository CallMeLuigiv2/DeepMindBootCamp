```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
sns.set_style('whitegrid')
```

## 1. Load Results

```python
with open('../results.json') as f:
    results = json.load(f)

# Display as table
for model, metrics in results.items():
    print(f"\n{model}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value}")
```

## 2. Model Comparison Bar Chart

```python
# Parse results into arrays for plotting
model_names = list(results.keys())
metrics_to_plot = ['accuracy', 'f1', 'roc_auc']

fig, axes = plt.subplots(1, len(metrics_to_plot), figsize=(5 * len(metrics_to_plot), 6))

for ax, metric in zip(axes, metrics_to_plot):
    means = []
    stds = []
    for model in model_names:
        val = results[model].get(metric, '0.0 +/- 0.0')
        if isinstance(val, str) and '+/-' in val:
            parts = val.split('+/-')
            means.append(float(parts[0].strip()))
            stds.append(float(parts[1].strip()))
        else:
            means.append(0.0)
            stds.append(0.0)

    bars = ax.barh(model_names, means, xerr=stds, capsize=3, color=sns.color_palette('viridis', len(model_names)))
    ax.set_xlabel(metric.replace('_', ' ').title())
    ax.set_title(f'{metric.replace("_", " ").title()} Comparison')
    ax.set_xlim(0, 1)

plt.tight_layout()
plt.savefig('../figures/model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```
