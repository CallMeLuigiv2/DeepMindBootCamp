## 1. Positional Encoding Visualization

```python
# Visualize the positional encoding matrix
pe = PositionalEncoding(d_model=64, max_len=50, dropout=0.0)
pe_matrix = pe.pe[0].numpy()  # (50, 64)
plot_positional_encoding(pe_matrix)

# Relative position test: show that dot product depends on offset, not absolute position
pe_large = PositionalEncoding(d_model=512, max_len=100, dropout=0.0)
pe_vecs = pe_large.pe[0]  # (100, 512)

# Compute dot products for different offsets
offsets = list(range(0, 20))
for pos in [0, 10, 30, 50]:
    dots = [torch.dot(pe_vecs[pos], pe_vecs[pos + k]).item() for k in offsets if pos + k < 100]
    plt.plot(offsets[:len(dots)], dots, label=f'pos={pos}')
plt.xlabel('Offset k')
plt.ylabel('Dot product PE(pos) . PE(pos+k)')
plt.title('Relative Position Test: Dot product depends on offset, not position')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
